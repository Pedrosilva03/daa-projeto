```python
'''def remove_duplicated_columns(df_train, df_test):
    # Transpõe o DataFrame para verificar duplicações de colunas com base no conteúdo
    duplicated_columns = df_train.T.duplicated(keep=False)
    
    # Filtra apenas as colunas duplicadas e seus conteúdos
    duplicated_data = df_train.loc[:, duplicated_columns]
    
    # Agrupa as colunas duplicadas com base no conteúdo
    grouped_duplicates = duplicated_data.T.groupby(list(duplicated_data.T)).groups

    columns_to_remove = []
    # Loop por cada grupo de colunas duplicadas, mantendo apenas a primeira e removendo as demais
    for content, columns in grouped_duplicates.items():
        columns_to_remove.extend(columns[1:])  # Mantém a primeira coluna e remove as demais

    # Remove colunas duplicadas do dataset de treino
    df_train.drop(columns=columns_to_remove, inplace=True)

    # Se um DataFrame de teste for fornecido, remova as colunas duplicadas que existem no teste
    if df_test is not None:
        df_test.drop(columns=[col for col in columns_to_remove if col in df_test.columns], inplace=True)

    # Contagem total de colunas duplicadas removidas
    total_duplicated_columns = len(columns_to_remove)
    
    print(f"Total de colunas duplicadas removidas: {total_duplicated_columns}")
    
    # Retorna os DataFrames e a contagem de colunas duplicadas
    return df_train, df_test, total_duplicated_columns

# Uso da função
df_train, df_test, total_duplicated_columns = remove_duplicated_columns(df_train, df_test)
'''

def remove_duplicated_columns(df):
    # Transpõe o DataFrame para identificar duplicações de colunas com base no conteúdo
    duplicated_columns = df.T.duplicated(keep=False)
    
    # Filtra apenas as colunas duplicadas e seus conteúdos
    duplicated_data = df.loc[:, duplicated_columns]
    
    # Agrupa as colunas duplicadas com base no conteúdo
    grouped_duplicates = duplicated_data.T.groupby(list(duplicated_data.T)).groups

    columns_to_remove = []
    # Loop por cada grupo de colunas duplicadas, mantendo apenas a primeira e removendo as demais
    for content, columns in grouped_duplicates.items():
        columns_to_remove.extend(columns[1:])  # Mantém a primeira coluna e remove as demais

    # Remove as colunas duplicadas do dataset
    df.drop(columns=columns_to_remove, inplace=True)

    # Contagem total de colunas duplicadas removidas
    total_duplicated_columns = len(columns_to_remove)
    
    print(f"Total de colunas duplicadas removidas: {total_duplicated_columns}")
    
    # Retorna o DataFrame modificado e a contagem de colunas duplicadas
    return df, total_duplicated_columns

# Uso da função
df, total_duplicated_columns = remove_duplicated_columns(df)
```

### Tratamento de Outliers

```python
# Copiando o dataframe original para preservar dados antes da transformação
df_tratado = df.copy()

# Loop pelas colunas para calcular IQR e ajustar os outliers
for coluna in df_tratado.select_dtypes(include=['float', 'int']).columns:  # Apenas colunas numéricas
    # Calcula Q1 e Q3
    Q1 = df_tratado[coluna].quantile(0.25)
    Q3 = df_tratado[coluna].quantile(0.75)
    IQR = Q3 - Q1
    
    # Limites para detecção de outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Tratar os outliers ajustando-os aos limites
    df_tratado[coluna] = df_tratado[coluna].apply(lambda x: limite_inferior if x < limite_inferior else (limite_superior if x > limite_superior else x))

# Verificação dos ajustes
print("Outliers tratados!")
df_tratado.describe()
```
